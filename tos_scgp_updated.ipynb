```python
import scanpy as sc
from gensim.corpora import Dictionary
import logging
from scgpt.model import TransformerModel
from scgpt.tokenizer import random_mask_value
import torch
from torch.utils.data import DataLoader, TensorDataset
from sklearn.metrics import accuracy_score, f1_score
import pandas as pd
import numpy as np
from anndata import AnnData
import scipy.sparse
import TOSICA
import os
os.environ["PYTORCH_SYMBOLIC_SHAPE_INFERENCE"] = "0"

# Placeholder logger setup for gene_tokenizer compatibility
logging.basicConfig(level=logging.INFO)
logger = logging.getLogger("gene_tokenizer")

# Tokenizer function (from gene_tokenizer.py) - modified to use placeholder logger
def tokenize_and_pad_batch(
    data: np.ndarray,
    gene_ids: np.ndarray,
    max_len: int,
    cls_id: int = 0,
    pad_id: int = 0,
    pad_value: int = 0,
    append_cls: bool = True,
    include_zero_gene: bool = False,
    return_pt: bool = True,
):
    tokenized_data = []
    for row in data:
        idx = np.nonzero(row)[0] if not include_zero_gene else range(len(row))
        values = row[idx]
        genes = gene_ids[idx]
        if append_cls:
            genes = np.insert(genes, 0, cls_id)
            values = np.insert(values, 0, 0)
        if return_pt:
            genes = torch.tensor(genes, dtype=torch.long)
            values = torch.tensor(values, dtype=torch.float)
        tokenized_data.append((genes, values))
    return tokenized_data

# Load the datasets 
train_dataset_path = "C:/Users/gaiacronus/Downloads/work/combine/adata/hPancreas_train_adata.h5ad"  
test_dataset_path = "C:/Users/gaiacronus/Downloads/work/combine/adata/hPancreas_test_adata.h5ad"  

adata_train = sc.read_h5ad(train_dataset_path)
adata_test = sc.read_h5ad(test_dataset_path)

# Step 1: Preprocess the datasets for TOSICA
# Preprocessing for the training dataset
sc.pp.normalize_total(adata_train, target_sum=1e4)
sc.pp.log1p(adata_train)
sc.pp.highly_variable_genes(adata_train, n_top_genes=2000, subset=True)

# Preprocessing for the test dataset
sc.pp.normalize_total(adata_test, target_sum=1e4)
sc.pp.log1p(adata_test)
sc.pp.highly_variable_genes(adata_test, n_top_genes=2000, subset=True)

# Step 2: Create vocabulary using gensim
# Combine genes from both datasets
all_genes = list(set(adata_train.var_names.tolist() + adata_test.var_names.tolist()))

# Add special tokens if needed (e.g., <pad>, <cls>)
special_tokens = ["<pad>", "<cls>"]
all_genes = special_tokens + all_genes
gene_vocab = GeneVocab(gene_list=gene_names, specials=special_tokens, default_token="<pad>")
#gene_vocab = Dictionary([all_genes])
#gene_vocab.add_documents([["<pad>"]])
# Step 3: Tokenize datasets using the vocabulary
def tokenize_data(adata, vocab, max_len=512):
    gene_ids = [vocab.token2id[gene] if gene in vocab.token2id else vocab.token2id["<pad>"] for gene in adata.var_names]
    #data_matrix = adata.X if not sc.sparse.issparse(adata.X) else adata.X.toarray()
    if isinstance(adata, AnnData):
        if scipy.sparse.issparse(adata.X):  # Check if sparse
            data_matrix = adata.X.toarray()  # Convert sparse to dense
        else:
            data_matrix = adata.X  # It's already a dense numpy array
    else:
        raise ValueError("Expected an AnnData object.")
    return tokenize_and_pad_batch(
        data=data_matrix,
        gene_ids=np.array(gene_ids),
        max_len=max_len,
        cls_id=vocab.token2id["<cls>"],
        pad_id=vocab.token2id["<pad>"],
        pad_value=0,
        append_cls=True,
        include_zero_gene=False,
        return_pt=True
    )

train_tokenized = tokenize_data(adata_train, gene_vocab)
test_tokenized = tokenize_data(adata_test, gene_vocab)

# Step 4: Apply SCGPT model
class SCGPTModel:
    def __init__(self, vocab_size, emb_size=128, num_heads=4, num_layers=4, d_hid=512, dropout=0.2):
        self.model = TransformerModel(vocab_size, emb_size, num_heads, num_layers, d_hid, dropout=dropout)

            
        

    def forward(self, tokenized_data):
        input_gene_ids = tokenized_data["genes"]
        input_values = random_mask_value(
            tokenized_data["values"], mask_ratio=0.15, mask_value=0, pad_value=0
        )

        src_key_padding_mask = input_gene_ids.eq(gene_vocab.token2id["<pad>"])

        with torch.no_grad():
            outputs = self.model(input_gene_ids, input_values, src_key_padding_mask)
        return outputs

# Initialize and apply the SCGPT model
scgpt_model = SCGPTModel(vocab_size=len(gene_vocab),vocab=gene_vocab)
scgpt_train_outputs = scgpt_model.forward(train_tokenized)
scgpt_test_outputs = scgpt_model.forward(test_tokenized)

# Step 5: Create DataLoader for evaluation
def create_data_loader(tokenized_data, batch_size=32):
    inputs = tokenized_data['genes']
    labels = tokenized_data['labels']  # Ensure your tokenized data includes labels
    dataset = TensorDataset(inputs, labels)
    return DataLoader(dataset, batch_size=batch_size)

train_loader = create_data_loader(train_tokenized)
test_loader = create_data_loader(test_tokenized)

# Step 6: Evaluation logic for SCGPT
def evaluate(model, data_loader, device):
    model.eval()
    all_preds = []
    all_labels = []
    with torch.no_grad():
        for batch in data_loader:
            inputs = batch[0].to(device)
            labels = batch[1].to(device)
            outputs = model.model(inputs)
            _, preds = torch.max(outputs, dim=1)
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())
    accuracy = accuracy_score(all_labels, all_preds)
    f1 = f1_score(all_labels, all_preds, average='macro')
    return accuracy, f1

# Perform evaluation
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
scgpt_model.model.to(device)

train_accuracy, train_f1 = evaluate(scgpt_model, train_loader, device)
test_accuracy, test_f1 = evaluate(scgpt_model, test_loader, device)

# Step 7: Save results to CSV
results = pd.DataFrame({
    "Dataset": ["Train", "Test"],
    "Accuracy": [train_accuracy, test_accuracy],
    "F1 Score": [train_f1, test_f1]
})

results.to_csv("results.csv", index=False)
print("Results saved to results.csv")

import scanpy as sc
import TOSICA
import os


# Load the h_pancreas datasets
train_path = "C:/Users/gaiacronus/Downloads/work/combine/adata/hPancreas_train_adata.h5ad"  # Training dataset
test_path = "C:/Users/gaiacronus/Downloads/work/combine/adata/hPancreas_test_adata.h5ad"   # Testing dataset

# Load the datasets as AnnData objects
train_adata = sc.read_h5ad(train_path)
test_adata = sc.read_h5ad(test_path)

# Preprocess the datasets
# Align the gene names in both datasets
common_genes = train_adata.var_names.intersection(test_adata.var_names)
train_adata = train_adata[:, common_genes]
test_adata = test_adata[:, common_genes]

# Convert AnnData object to scGPT-compatible format
train_scGPT = prepare_for_training(train_adata)
test_scGPT = prepare_for_training(test_adata)

# Normalize the data
sc.pp.normalize_total(train_adata, target_sum=1e4)
sc.pp.log1p(train_adata)
sc.pp.normalize_total(test_adata, target_sum=1e4)
sc.pp.log1p(test_adata)

ref_adata = train_adata
query_adata = test_adata
label_key = "Celltype2"
gmt_path = "C:/Users/gaiacronus/Downloads/work/combine/TOSICA/resources/immune.gmt"  # Path to the gene set file
project_name = "tostry"

# Train the model
TOSICA.train(
    ref_adata,
    gmt_path=gmt_path,
    project=project_name,
    epochs=3,
    label_name=label_key
)

# Predict cell types in the test dataset
model_weight_path = f"{project_name}/model_weight.h5"

# Loading pre-trained scGPT model and fine-tune
model = load_pretrained_scGPT("best_model.pt")

# Fine-tune the model using the training data
model.train(train_scGPT, epochs=3, batch_size=32)

# Predict cell types
predicted_adata = TOSICA.pre(
    query_adata,
    model_weight_path=model_weight_path,
    project=project_name
)

predictions = model.predict(test_scGPT)
test_adata.obs["scGPT_predictions"] = predictions


# Predicted cell types and probabilities are added to the `obs` of the AnnData object
print(predicted_adata.obs.head())  # Check the predicted labels and probabilities

# Save the results for further analysis
predicted_adata.write("h_pancreas_predicted.h5ad")

# Compare TOSICA and scGPT results
print("TOSICA Predictions:\n", predicted_adata.obs.head())
print("scGPT Predictions:\n", test_adata.obs["scGPT_predictions"].head())
```
